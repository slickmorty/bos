```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams["figure.figsize"]= (12,8)

indicator_images_path = "D:/dev/workspace/vscode/python/bos/data/indicators/images"


def plot_with_close(df: pd.DataFrame, name: str, *,
                    interval: int = 0,
                    list_of_random_numbers: list[int] = None,
                    save: bool = False):
    if save:

        for index, random in enumerate(list_of_random_numbers):
            fig, axs = plt.subplots(2)
            fig.patch.set_facecolor('white')
            
            axs[0].set_title(name)
            axs[0].plot(df[name][random:random+interval])
            axs[1].set_title("Close")
            axs[1].plot(df.Close[random:random+interval])

            # plt.legend()
            plt.savefig(f"{indicator_images_path}/{name}_{index+1}.png")
            plt.close()
        return


data_name = "SP500m_H1_202009231400_202206172200"
csv_path = "D:/dev/workspace/vscode/python/bos/data/csv/"
data_indicator_path = csv_path+"with_indicator/" + data_name + ".csv"
df = pd.read_csv(data_indicator_path)
df.pop("DateTime")

interval = 500
random_list = [np.random.randint(0, len(df)-interval) for _ in range(25)]

for column in df.columns:
    plot_with_close(df,
                    column,
                    interval=interval,
                    list_of_random_numbers=random_list,
                    save=True)
```
